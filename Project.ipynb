```python
#2
filepath=r'C:\Users\Rupali\Datafiles\AppleStore.csv'
```

```python
df = pd.read_csv(filepath)
df
```

```python
## 2.1
t=df[['prime_genre' , 'user_rating']]
g=t[t.prime_genre == 'Games']
ng=t[t.prime_genre != 'Games']
print('Mean{0}'.format(g.mean()))
print('Mean non games{0}'.format(ng.mean()))
```

```python
games=['Free apps','Affordable apps','Expensive apps','Very expensive']
rating1=df[df.price==0]
rating2=df[(df.price>0)&(df.price<20)]
rating3=df[(df.price>20)&(df.price<50)]
rating4=df[df.price>50]
free=rating1.count()[0]
af_ap=rating2.count()[0]
ea=rating3.count()[0]
ve=rating4.count()[0]
li=[free,af_ap,ea,ve]
plt.pie(li,labels=games,shadow =True,startangle=90,autopct='%1.1f%%')
plt.legend(loc='best')
plt.subplots_adjust(0.3,0,2,2)
plt.show()
```
